# Exporting vector database

This notebook serves as a script to run in Google Colab to load the full dataset and export it into a file, which we can then store in repo and load whenever we like. This will significantly shorten initialisation time of our pipeline

```python
rag = RAG()

rag.load_embedding_model()
```

```python
books_paths = []
data_root_path = 'NLP_RAG/data'

for filename in os.listdir(data_root_path):
    books_paths.append(os.path.join(data_root_path, filename))

books_paths
```

```python
rag.add_books(books_paths)
```

```python
rag.save_vector_store('books_default')
```
